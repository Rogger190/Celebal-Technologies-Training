```python
# Python Operators
# Arithmetic Operators
a = int(input())
b = int(input())

print("Addition Operator: a + b =",a+b)
print("Substraction Operator: a - b =",a-b)
print("Multiplication Operator: a * b =",a*b)
print("Modulus Operator: a % b =",a%b)
print("Integer Division Operator: a / b =",a/b)
print("Float Division Operator: a + b =",a//b)
print("Exponentiation Operator: a ** b =",a**b)
```

```python
# Logical Operators
x = bool(input())
y = bool(input())

print("AND Operator: x AND y:",x and y)
print("OR Operator: x OR y:",x or y)
print("NOT Operator: NOT x:", not x)
```

```python
# Relational/Comparision Operators
p = int(input())
q = int(input())

print("p < q Operator:",p<q)
print("p > q Operator:",p>q)
print("p <= q Operator:",p<=q)
print("p >= q Operator:",p>=q)
print("p == q Operator:",p==q)
print("p != q Operator:",p!=q)
```

```python
# Assignment Operators
m = int(input())
n = int(input())

print("= Operator: m = n",m=n)
print("+= Operator: m += n",m+=n)
print("-= Operator: m -= n",m-=n)
print("*= Operator: m*= n",m*=n)
print("/= Operator: m /= n",m/=n)
print("%= Operator: m %= n",m%=n)
print("//= Operator: m //= n",m//=n)
print("**= Operator: m **= n",m**=n)
print("&= Operator: m &= n",m&=n)
print("|= Operator: m |= n",m|=n)
print("^= Operator: m ^= n",m^=n)
```

```python
# Identity Operators
# is Operator
x = ["pratik", "sameer"]
y = ["pratik", "sameer"]
z = x

print(x is z)

# returns True because z is the same object as x

print(x is y)

# returns False because x is not the same object as y, even if they have the same content

# is not Operator
x = ["pratik", "sameer"]
y = ["pratik", "sameer"]
z = x

print(x is not z)

# returns False because z is the same object as x

print(x is not y)

# returns True because x is not the same object as y, even if they have the same content

```

```python
# Membership Operators
# in Operator
x = ["pratik", "sameer"]

print("sameer" in x)

# returns True because a sequence with the value "banana" is in the list

# not in Operator
x = ["pratik", "sameer"]

print("pineapple" not in x)

# returns True because a sequence with the value "pineapple" is not in the list
```

```python
# Bitwise Operators

g = int(input())
h = int(input())

print("AND Operator: g & h =",g & h)
print("OR Operator: g | h =",g | h)
print("XOR Operator: g ^ h =",g ^ h)
print("LEFT SHIFT Operator: g << h =",g << h)
print("LEFT SHIFT Operator: g << h =",g << h)
print("Right SHIFT Operator: g >> h =",g >> h)

```
